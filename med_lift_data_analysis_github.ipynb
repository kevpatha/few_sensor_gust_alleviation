```python
def diff_mean_and_std(diff, num_seeds=None):
    if num_seeds:
        steps = np.shape(diff)[0]
        mean_diff = np.zeros((steps, num_seeds))
        std_diff = np.zeros((steps, num_seeds))
        for i in range(num_seeds):
            ind = i*10
            mean_diff[:,i] = np.mean(diff[:,ind:(ind+num_seeds)], axis=1)
            std_diff[:,i] = np.std(diff[:,ind:(ind+num_seeds)], axis=1)
        
        
        diff_mean = np.mean(mean_diff,axis=1)
        diff_std = np.sum(std_diff**2,axis=1)
        diff_std = np.sqrt(diff_std)/(num_seeds)
   
        return mean_diff, std_diff, diff_mean, diff_std
    else:
        return np.mean(diff, axis=1), np.std(diff, axis=1)
            
```

```python
def median_confidence(data, n=5, q=0.5):
    j = n*q-1.96*np.sqrt(n*q*(1-q))
    k = n*q+1.96*np.sqrt(n*q*(1-q))
    
    #print("JK",j,k)
    data_sorted = np.sort(data, axis=0)
    top = data_sorted[2]#[int(k)] #same as rounding up and then subtracting one since python [0] is first
    bot = data_sorted[0]#[int(j)-1]
    return bot,top
```

```python
from scipy import stats

def check_outliers_median(avg, avg_tot):
    #checking for outliers
    
    Q1 = np.quantile(avg_tot, 0.25)
    Q3 = np.quantile(avg_tot, 0.75)
    iqr = Q3-Q1
    out_cap = Q3+1.5*iqr
    #print(Q1, Q3)
    #print('outlier cap: ', out_cap)
    med_avg_tot = np.median(avg_tot)
    bot_conf, top_conf = median_confidence(avg_tot)
    max_ind = np.argmax(avg_tot)
    mean_avg = np.mean(avg, axis=1)
    mean_avg_tot = np.mean(avg_tot)
    std_avg_tot = np.std(avg_tot)
    if avg_tot[max_ind]> out_cap:
        #print('max_ind: ', max_ind)
        #print('dist away from out_cap: ', avg_tot[max_ind]-out_cap)
        
        #avg = np.delete(avg,max_ind,1)
        #avg_tot = np.mean(avg[100:,:], axis=0)
        avg_tot[max_ind] = out_cap
        med_avg_tot = np.median(avg_tot)
        _, top_conf = median_confidence(avg_tot, n=len(avg_tot))
        mean_avg = np.mean(avg, axis=1)
    else:
        print('none')
        
    return mean_avg, avg_tot, mean_avg_tot, std_avg_tot, top_conf
```

```python
def error_reduct(base, rl, ax, row, col, max_row, xlabel=None, ylabel=None, title=None):
    
    avg_err_reduct1, avg_total_err_reduct1 = get_avg_tot_err_red(base,rl)
    #mean_avg_total_err_reduct1 = np.mean(avg_total_err_reduct1)
    #top_conf_err_reduct1 = np.std(avg_total_err_reduct1)
    
    #checking for outliers
    print('row,col: ', row,col)
    avg, avg_total_err_reduct1, mean_avg_total_err_reduct1, std_avg_tot_err_reduct1, top_conf_err_reduct1 = check_outliers_median(avg_err_reduct1,
                                                                                                                              avg_total_err_reduct1)
    
    
    x = range(len(avg_err_reduct1))
    ax[row,col].plot(avg_err_reduct1)
    #ax[row,col].fill_between(x,avg_err_reduct1-std_err_reduct1*1.96,avg_err_reduct1+std_err_reduct1*1.96, alpha=0.2, color='tab:blue')
    ax[row,col].plot(x,np.ones((len(x),len(avg_total_err_reduct1)))*avg_total_err_reduct1, 'k--')
    #ax[row,col].plot(x,np.ones(len(x),)*top_conf_err_reduct1, 'r--')
    #ax[row,col].plot(x,np.ones(len(x),)*out_cap, 'g--')
    #ax[0].plot(x,np.ones(len(x),)*(med_avg_err_reduct1-std_avg_err_reduct1*1.96), 'r--')
    ax[row,col].set_ylim(0,1)
    
    ax[row,col].spines["top"].set_visible(False)
    ax[row,col].spines["right"].set_visible(False)
    
    ax[row,col].set_ylim(0, 1)
    ax[row,col].set_xlim(-1,100)
    
    if col > 0:
        ax[row,col].spines["left"].set_visible(False)
        ax[row,col].yaxis.set_visible(False)
        
    if row < max_row-1:
        ax[row,col].spines["bottom"].set_visible(False)
        ax[row,col].xaxis.set_visible(False)

    if xlabel:
        ax[row,col].set_xlabel(xlabel)
    if ylabel:
        ax[row,col].set_ylabel(ylabel)
    if title:
        ax[row,col].set_title(title)
    
    
        
    return np.array([mean_avg_total_err_reduct1, std_avg_tot_err_reduct1])
```

```python
from scipy import stats

def check_median_iqr(avg, avg_tot):
    #checking for outliers
    
    Q1 = np.quantile(avg_tot, 0.25)
    Q3 = np.quantile(avg_tot, 0.75)
    iqr = Q3-Q1
    out_cap = Q3+1.5*iqr
    #print(Q1, Q3)
    #print('outlier cap: ', out_cap)
    med_avg_tot = np.median(avg_tot)
    max_ind = np.argmax(avg_tot)
    mean_avg = np.mean(avg, axis=1)
        
    return mean_avg, avg_tot, med_avg_tot, Q1, Q3
```

```python
def get_rise_time(time, rl):
    gust_err = rl[51:147,:]
    avg_total_reduced_err = np.mean(gust_err[-49:,:], axis=0)
    start_err = np.zeros(np.shape(gust_err)[1],)
    for k in range(np.shape(gust_err)[1]):
        if gust_err[1,k] < 0:
            start_err[k] = np.min(gust_err[0:10,k])
        else:
            start_err[k] = np.max(gust_err[0:10,k], axis = 0)
    tot_err_diff = start_err - avg_total_reduced_err
    r_time = np.zeros((50,))
    r_start = np.zeros((50))
    need_r_start = True
    
    for ii in range(np.shape(gust_err)[1]):
        for i in range(np.shape(gust_err)[0]):
            if start_err[ii] < 0:
                if need_r_start:
                    if start_err[ii] - gust_err[i,ii] <= tot_err_diff[ii]*0.1:
                        r_start[ii] = time[i]
                        need_r_start = False
                if start_err[ii] - gust_err[i,ii] <= tot_err_diff[ii]*0.9:
                    r_time[ii] = time[i]-r_start[ii]
                    need_r_start = True
                    break
            else:
                if need_r_start:
                    if start_err[ii] - gust_err[i,ii] >= tot_err_diff[ii]*0.1:
                        r_start[ii] = time[i]
                        need_r_start = False
                if start_err[ii] - gust_err[i,ii] >= tot_err_diff[ii]*0.9:
                    r_time[ii] = time[i]-r_start[ii]
                    need_r_start = True
                    break
    
    
    return r_time
```
